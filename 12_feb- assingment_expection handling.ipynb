```python
Question: What happens when an exception is not handled? explain with an example.

Answer: if the exception not handled, they will terminate the program and produce a traceback.

Eample: Exception Handling Using try... except
try:
    numerator = 10
    denominator= 0
    
    result = numerator/denominator
    
    print(result)
except:
    print("error: denominator cannot be 0.")
#output: Error: Denominator cannot be 0.

in this example, we are trying to divide a number by 0.
here, this code generates an exception.

To handle the exception, we have put the code,
[result = numerator/denominator] inside the [try] block. now when an exception occurs, the rest of the code inside the [try] block is skipped.
the [except] block cathces the exception and statements inside the [except] block are executed.
```

```python
Question: What are custom exceptions in python? why do we need custom exceptions? explain with an example.
Answer: 
Defining custom exceptions:
Like standard exception classes, custom exceptions are also classes. hence, you can add functionality to the custom exception classes like:
~ Adding attribute and properties.
~adding methods e.g., log the exception, format the output, etc.

Why do we need exceptions:
custom exceptions provide you the flexibility to add attribute and methods that are not part of a standard exception. these can store additional information.

Example:
class NegativeError(exception):
    def __init__(self, data):
        self.data = data
try:
    x = int(input("Enter a number between positive integer:"))
    if x<0:
        raise NegativeError(x)
except NegativeError as e:
    print("you provided {}. please provide positive integer values only". format (e))
```
